```python
# í•¨ìˆ˜ ë§Œë“¤ê¸°: get_top_places
def get_top_places(region_name, category_name, top_n=5):
    category_code = category_code_map.get(category_name)
    if not category_code:
        print(f"âŒ '{category_name}' ì¹´í…Œê³ ë¦¬ ì½”ë“œê°€ ì—†ìŠµë‹ˆë‹¤.")
        return []

    category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

    q = f"""
    PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ktop: <http://data.visitkorea.or.kr/property/>

    SELECT ?name
    WHERE {{
      ?place a kto:Place ;
             rdfs:label ?name ;
             ktop:category <{category_uri}> .
      FILTER(CONTAINS(?name, "{region_name}"))
    }}
    """

    results = g.query(q)

    # ì´ë¦„ ë¦¬ìŠ¤íŠ¸ë¡œ ëª¨ìœ¼ê¸°
    names = [str(row.name) for row in results]

    # ì´ë¦„ ê¸¸ì´ ìˆœ ì •ë ¬ (ëŒ€ì¶© ìœ ëª…í• ìˆ˜ë¡ ì´ë¦„ì´ ê¸¸ê±°ë‚˜ "êµ­ë¦½" ê°™ì€ í‚¤ì›Œë“œ í¬í•¨í•  ìˆ˜ ìˆìŒ)
    names_sorted = sorted(names, key=lambda x: (-("êµ­ë¦½" in x), len(x)))

    return names_sorted[:top_n]
```

```python
# ì‚¬ìš©ë²• ì˜ˆì œ - ì„œìš¸ì— ìˆëŠ” ë°•ë¬¼ê´€ TOP 5
top_places = get_top_places("ì„œìš¸", "ë°•ë¬¼ê´€", top_n=5)

print("âœ… ì„œìš¸ ë°•ë¬¼ê´€ TOP 5:")
for idx, name in enumerate(top_places, 1):
    print(f"{idx}. {name}")
```

```python
# "êµ­ë¦½", "ì‹œë¦½", "ì„¸ê³„ë¬¸í™”ìœ ì‚°" í‚¤ì›Œë“œ ìš°ì„  ì •ë ¬ í•¨ìˆ˜
def get_top_places(region_name, category_name, top_n=5):
    # ğŸ”¹ ìš°ì„ ìˆœìœ„ í‚¤ì›Œë“œ ë¦¬ìŠ¤íŠ¸
    priority_keywords = ["êµ­ë¦½", "ì‹œë¦½", "ì„¸ê³„ë¬¸í™”ìœ ì‚°"]

    # ì¹´í…Œê³ ë¦¬ ì½”ë“œ ê°€ì ¸ì˜¤ê¸°
    category_code = category_code_map.get(category_name)
    if not category_code:
        print(f"âŒ '{category_name}' ì¹´í…Œê³ ë¦¬ ì½”ë“œê°€ ì—†ìŠµë‹ˆë‹¤.")
        return []

    category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

    # SPARQL ì¿¼ë¦¬
    q = f"""
    PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ktop: <http://data.visitkorea.or.kr/property/>

    SELECT ?name
    WHERE {{
      ?place a kto:Place ;
             rdfs:label ?name ;
             ktop:category <{category_uri}> .
      FILTER(CONTAINS(?name, "{region_name}"))
    }}
    """

    results = g.query(q)

    # ì´ë¦„ ë¦¬ìŠ¤íŠ¸ë¡œ ëª¨ìœ¼ê¸°
    names = [str(row.name) for row in results]

    # ğŸ¯ ì •ë ¬ ê¸°ì¤€:
    # 1. priority_keywords ì¤‘ ëª‡ ê°œë‚˜ í¬í•¨ëëŠ”ì§€ (ë§ì„ìˆ˜ë¡ ìš°ì„ )
    # 2. ì´ë¦„ ê¸¸ì´ ê¸´ ìˆœ

    def priority_score(name):
        return sum(keyword in name for keyword in priority_keywords)

    names_sorted = sorted(
        names,
        key=lambda x: (-priority_score(x), -len(x))  # í‚¤ì›Œë“œ ë§ì´ í¬í•¨í• ìˆ˜ë¡ ìš°ì„ , ê·¸ë‹¤ìŒ ì´ë¦„ ê¸¸ì´
    )

    return names_sorted[:top_n]
```

```python
# ì‚¬ìš© ì˜ˆì‹œ
# í¬í•¨ í‚¤ì›Œë“œ ìˆ˜ê°€ ê°™ìœ¼ë©´ ì´ë¦„ ê¸¸ì´ ê¸´ ìˆœì„œë¡œ
top_places = get_top_places("ì„œìš¸", "ë°•ë¬¼ê´€", top_n=5)

print("âœ… ì„œìš¸ ë°•ë¬¼ê´€ TOP 5 (êµ­ë¦½, ì‹œë¦½, ì„¸ê³„ë¬¸í™”ìœ ì‚° ìš°ì„ ):")
for idx, name in enumerate(top_places, 1):
    print(f"{idx}. {name}")
```

```python
import pandas as pd
import re

# âœ… ì§€ì—­ ë¦¬ìŠ¤íŠ¸
regions = ["ì„œìš¸", "ë¶€ì‚°", "ì œì£¼", "ê°•ì›", "ê²½ê¸°", "ì „ë¼", "ê²½ìƒ", "ì¶©ì²­", "ì¸ì²œ", "ëŒ€ì „", "ê´‘ì£¼", "ìš¸ì‚°", "ì„¸ì¢…"]

# âœ… ìœ í˜•ëª… â†’ RDF category code ë§¤í•‘
category_code_map = {
    "ë°•ë¬¼ê´€": "A04011000",
    "ë¯¸ìˆ ê´€": "A04011200",
    "í•´ë³€": "A01020100",
    "ì‚°ì„±": "A03010200",
    "ì‚¬ì°°": "A03010300",
    "ë†ì´Œì²´í—˜": "A02081400",
    "ë¦¬ì¡°íŠ¸": "B02011600",
}

# âœ… ë°ì´í„° ìˆ˜ì§‘
all_data = []

for region in regions:
    for category_name, category_code in category_code_map.items():
        category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

        # SPARQL ì¿¼ë¦¬ (ì˜ˆì•½ URL, ì „í™”ë²ˆí˜¸ ë“± ë‹¤ì–‘í•œ í•„ë“œ í¬í•¨)
        q = f"""
        PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
        PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
        PREFIX ktop: <http://data.visitkorea.or.kr/property/>

        SELECT ?name ?reservation ?homepage ?phone ?tel ?contact
        WHERE {{
            ?place a kto:Place ;
                   rdfs:label ?name ;
                   ktop:category <{category_uri}> .
            FILTER(CONTAINS(?name, "{region}"))
            OPTIONAL {{ ?place ktop:reservation ?reservation }}
            OPTIONAL {{ ?place ktop:homepage ?homepage }}
            OPTIONAL {{ ?place ktop:serviceUrl ?homepage }}
            OPTIONAL {{ ?place ktop:phoneNumber ?phone }}
            OPTIONAL {{ ?place ktop:tel ?tel }}
            OPTIONAL {{ ?place ktop:contact ?contact }}
        }}
        """

        for row in g.query(q):
            reservation_url = (
                str(row.reservation) if row.reservation else
                str(row.homepage) if row.homepage else ""
            )

            phone_number = (
                str(row.phone) if row.phone else
                str(row.tel) if row.tel else
                str(row.contact) if row.contact else ""
            )

            all_data.append({
                "ì§€ì—­": region,
                "ìœ í˜•": category_name,
                "ê´€ê´‘ì§€ëª…": str(row.name),
                "ì˜ˆì•½ URL": reservation_url,
                "ì „í™”ë²ˆí˜¸": phone_number
            })

# âœ… ë°ì´í„°í”„ë ˆì„ ìƒì„±
df_all = pd.DataFrame(all_data)

# âœ… ì „í™”ë²ˆí˜¸ ì •ê·œí™” í•¨ìˆ˜ ì •ì˜
def normalize_phone(phone):
    if not phone or phone.strip() == "":
        return ""
    cleaned = re.sub(r"[^\d]", "", phone)
    if cleaned.startswith("0"):
        return "+82" + cleaned[1:]
    return cleaned

# âœ… ì •ê·œí™”ëœ ì „í™”ë²ˆí˜¸ ì»¬ëŸ¼ ì¶”ê°€
df_all["ì „í™”ë²ˆí˜¸ (ì •ê·œí™”)"] = df_all["ì „í™”ë²ˆí˜¸"].apply(normalize_phone)

# âœ… ì˜ˆì•½ ê°€ëŠ¥í•œ ê´€ê´‘ì§€ í•„í„°ë§ (ì„ íƒ)
df_reservation = df_all[df_all["ì˜ˆì•½ URL"].str.strip() != ""]

# âœ… í†µê³„ ìš”ì•½
print("ğŸ“Œ ì „ì²´ ê´€ê´‘ì§€ ìˆ˜:", len(df_all))
print("ğŸ“Œ ì˜ˆì•½ ê°€ëŠ¥í•œ ê´€ê´‘ì§€ ìˆ˜:", len(df_reservation))
print("ğŸ“ ì „í™”ë²ˆí˜¸ê°€ ìˆëŠ” ê´€ê´‘ì§€ ìˆ˜:", len(df_all[df_all["ì „í™”ë²ˆí˜¸"].str.strip() != ""]))

# âœ… CSV ì €ì¥
df_all.to_csv("ìµœì¢…_ê´€ê´‘ì§€ì •ë³´_ì˜ˆì•½ì „í™”_ì •ê·œí™”.csv", index=False, encoding="utf-8-sig")
print("ğŸ’¾ ì €ì¥ ì™„ë£Œ: ìµœì¢…_ê´€ê´‘ì§€ì •ë³´_ì˜ˆì•½ì „í™”_ì •ê·œí™”.csv")
```
