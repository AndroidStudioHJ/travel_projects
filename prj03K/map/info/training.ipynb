```python
# 유형명 ↔ 코드 매핑 딕셔너리
category_code_map = {
    "박물관": "A04011000",
    "미술관": "A04011200",
    "해변": "A01020100",
    "산성": "A03010200",
    "사찰": "A03010300",
    "농촌체험": "A02081400",
    "리조트": "B02011600",
}
```

```python
# 지역 + 유형으로 SPARQL 실행하는 함수
def get_places_by_region_and_category(region_name, category_name):
    from rdflib import URIRef

    category_code = category_code_map.get(category_name)
    if not category_code:
        print(f"❌ '{category_name}'에 해당하는 카테고리 코드가 없습니다.")
        return

    category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

    q = f"""
    PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ktop: <http://data.visitkorea.or.kr/property/>

    SELECT ?name ?category
    WHERE {{
      ?place a kto:Place ;
             rdfs:label ?name ;
             ktop:category <{category_uri}> .

      FILTER(CONTAINS(?name, "{region_name}"))
    }}
    """

    results = g.query(q)

    print(f"✅ '{region_name}' 지역의 '{category_name}' 목록:")
    count = 0
    for row in results:
        print(f"🏞 {row.name}")
        count += 1

    if count == 0:
        print("🙁 결과가 없습니다.")
    else:
        print(f"\n🔎 총 {count}개 결과")
```

# 지역 + 유형별 TOP 관광지 뽑기

```python
# 관광지 유형명 ↔ 코드 매핑
category_code_map = {
    "박물관": "A04011000",
    "미술관": "A04011200",
    "해변": "A01020100",
    "산성": "A03010200",
    "사찰": "A03010300",
    "농촌체험": "A02081400",
    "리조트": "B02011600",
    "공연장": "A03010300",
}
```

```python
# 함수 만들기: get_top_places
def get_top_places(region_name, category_name, top_n=5):
    category_code = category_code_map.get(category_name)
    if not category_code:
        print(f"❌ '{category_name}' 카테고리 코드가 없습니다.")
        return []

    category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

    q = f"""
    PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ktop: <http://data.visitkorea.or.kr/property/>

    SELECT ?name
    WHERE {{
      ?place a kto:Place ;
             rdfs:label ?name ;
             ktop:category <{category_uri}> .
      FILTER(CONTAINS(?name, "{region_name}"))
    }}
    """

    results = g.query(q)

    # 이름 리스트로 모으기
    names = [str(row.name) for row in results]

    # 이름 길이 순 정렬 (대충 유명할수록 이름이 길거나 "국립" 같은 키워드 포함할 수 있음)
    names_sorted = sorted(names, key=lambda x: (-("국립" in x), len(x)))

    return names_sorted[:top_n]
```

```python
# 사용법 예제 - 서울에 있는 박물관 TOP 5
top_places = get_top_places("서울", "박물관", top_n=5)

print("✅ 서울 박물관 TOP 5:")
for idx, name in enumerate(top_places, 1):
    print(f"{idx}. {name}")
```

```python
# "국립", "시립", "세계문화유산" 키워드 우선 정렬 함수
def get_top_places(region_name, category_name, top_n=5):
    # 🔹 우선순위 키워드 리스트
    priority_keywords = ["국립", "시립", "세계문화유산"]

    # 카테고리 코드 가져오기
    category_code = category_code_map.get(category_name)
    if not category_code:
        print(f"❌ '{category_name}' 카테고리 코드가 없습니다.")
        return []

    category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

    # SPARQL 쿼리
    q = f"""
    PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    PREFIX ktop: <http://data.visitkorea.or.kr/property/>

    SELECT ?name
    WHERE {{
      ?place a kto:Place ;
             rdfs:label ?name ;
             ktop:category <{category_uri}> .
      FILTER(CONTAINS(?name, "{region_name}"))
    }}
    """

    results = g.query(q)

    # 이름 리스트로 모으기
    names = [str(row.name) for row in results]

    # 🎯 정렬 기준:
    # 1. priority_keywords 중 몇 개나 포함됐는지 (많을수록 우선)
    # 2. 이름 길이 긴 순

    def priority_score(name):
        return sum(keyword in name for keyword in priority_keywords)

    names_sorted = sorted(
        names,
        key=lambda x: (-priority_score(x), -len(x))  # 키워드 많이 포함할수록 우선, 그다음 이름 길이
    )

    return names_sorted[:top_n]
```

```python
# 사용 예시
# 포함 키워드 수가 같으면 이름 길이 긴 순서로
top_places = get_top_places("서울", "박물관", top_n=5)

print("✅ 서울 박물관 TOP 5 (국립, 시립, 세계문화유산 우선):")
for idx, name in enumerate(top_places, 1):
    print(f"{idx}. {name}")
```

#  최종 목표: 통합 테이블 구조

```python
# 기본 요소

# 지역 리스트
regions = ["서울", "부산", "제주", "강원", "경기", "전라", "경상", "충청", "인천", "대전", "광주", "울산", "세종"]
# 맵핑
category_code_map = {
    "박물관": "A04011000",
    "미술관": "A04011200",
    "해변": "A01020100",
    "산성": "A03010200",
    "사찰": "A03010300",
    "농촌체험": "A02081400",
    "리조트": "B02011600",
    "공연장": "A03010300",
}
```

```python
import pandas as pd
import re

# ✅ 지역 리스트
regions = ["서울", "부산", "제주", "강원", "경기", "전라", "경상", "충청", "인천", "대전", "광주", "울산", "세종"]

# ✅ 유형명 → RDF category code 매핑
category_code_map = {
    "박물관": "A04011000",
    "미술관": "A04011200",
    "해변": "A01020100",
    "산성": "A03010200",
    "사찰": "A03010300",
    "농촌체험": "A02081400",
    "리조트": "B02011600",
}

# ✅ 데이터 수집
all_data = []

for region in regions:
    for category_name, category_code in category_code_map.items():
        category_uri = f"http://data.visitkorea.or.kr/resource/{category_code}"

        # SPARQL 쿼리 (예약 URL, 전화번호 등 다양한 필드 포함)
        q = f"""
        PREFIX kto: <http://data.visitkorea.or.kr/ontology/>
        PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
        PREFIX ktop: <http://data.visitkorea.or.kr/property/>

        SELECT ?name ?reservation ?homepage ?phone ?tel ?contact
        WHERE {{
            ?place a kto:Place ;
                   rdfs:label ?name ;
                   ktop:category <{category_uri}> .
            FILTER(CONTAINS(?name, "{region}"))
            OPTIONAL {{ ?place ktop:reservation ?reservation }}
            OPTIONAL {{ ?place ktop:homepage ?homepage }}
            OPTIONAL {{ ?place ktop:serviceUrl ?homepage }}
            OPTIONAL {{ ?place ktop:phoneNumber ?phone }}
            OPTIONAL {{ ?place ktop:tel ?tel }}
            OPTIONAL {{ ?place ktop:contact ?contact }}
        }}
        """

        for row in g.query(q):
            reservation_url = (
                str(row.reservation) if row.reservation else
                str(row.homepage) if row.homepage else ""
            )

            phone_number = (
                str(row.phone) if row.phone else
                str(row.tel) if row.tel else
                str(row.contact) if row.contact else ""
            )

            all_data.append({
                "지역": region,
                "유형": category_name,
                "관광지명": str(row.name),
                "예약 URL": reservation_url,
                "전화번호": phone_number
            })

# ✅ 데이터프레임 생성
df_all = pd.DataFrame(all_data)

# ✅ 전화번호 정규화 함수 정의
def normalize_phone(phone):
    if not phone or phone.strip() == "":
        return ""
    cleaned = re.sub(r"[^\d]", "", phone)
    if cleaned.startswith("0"):
        return "+82" + cleaned[1:]
    return cleaned

# ✅ 정규화된 전화번호 컬럼 추가
df_all["전화번호 (정규화)"] = df_all["전화번호"].apply(normalize_phone)

# ✅ 예약 가능한 관광지 필터링 (선택)
df_reservation = df_all[df_all["예약 URL"].str.strip() != ""]

# ✅ 통계 요약
print("📌 전체 관광지 수:", len(df_all))
print("📌 예약 가능한 관광지 수:", len(df_reservation))
print("📞 전화번호가 있는 관광지 수:", len(df_all[df_all["전화번호"].str.strip() != ""]))

# ✅ CSV 저장
df_all.to_csv("최종_관광지정보_예약전화_정규화.csv", index=False, encoding="utf-8-sig")
print("💾 저장 완료: 최종_관광지정보_예약전화_정규화.csv")
```
